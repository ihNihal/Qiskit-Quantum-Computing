# Importing Libraries

```python
from qiskit import *
from qiskit.tools.monitor import job_monitor
from qiskit.tools.visualization import plot_state_qsphere
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
```

```python
# Create circuit
circuit = QuantumCircuit(1,1)

# apply gate
circuit.h(0)

#apply measurement
circuit.measure(0,0)

#circuit draw
circuit.draw(output = 'mpl', initial_state = True)
```

```python
# select simulator
simulator = Aer.get_backend('aer_simulator')

# Select statevector
circuit.save_statevector()

# assemble circuit
qobj = assemble(circuit)

job = simulator.run(qobj)

#monitor job
job_monitor(job)

# get the result
result = job.result()

# get statevector
final_state = result.get_statevector()
```

```python
print(final_state)
array_to_latex(final_state)
```

```python
count = result.get_counts()
plot_histogram(count)
```

```python
plot_bloch_multivector(final_state)
```

```python
plot_state_qsphere(final_state)
```

# Uncertainity principle in action

3 basis:
1. Z basis / standard basis / computational basis
2. X basis / hadamard basis
3. Y basis / circular basis


IF we initialize & measure in z basis -> the outcome is certain. If we initialize measrue in X basis -> the outcome is random

# z basis measurement


```python
 # create a circuit
 circuit = QuantumCircuit(1,1)

 # apply gate
circuit.h(0)

 # apply measurement
 circuit.measure(0,0) #z basis measurement

 # draw circuit
 circuit.draw(output = 'mpl', initial_state =True)
```

```python
# select simulator
simulator = Aer.get_backend('aer_simulator')

# Select statevector
circuit.save_statevector()

# assemble circuit
qobj = assemble(circuit)

job = simulator.run(qobj)

#monitor job
job_monitor(job)

# get the result
result = job.result()

# get statevector
final_state = result.get_statevector()
```

```python
plot_bloch_multivector(final_state)
```
