Altering the circuit with Y basis Measurement

```python
# Building Circuit
circuit = QuantumCircuit(3,1)

# apply gate
circuit.h(0)
circuit.h(1)
circuit.ccx(0,1,2)
circuit.sx(2)

# apply measurement in y  basis
circuit.sdg(0)
circuit.h(0)

circuit.measure(2,0)





# drawing circuit
circuit.draw(output='mpl', initial_state=True)
```

```python

# select simulator
simulator = Aer.get_backend('aer_simulator')

# Select statevector
circuit.save_statevector()

# assemble circuit
qobj = assemble(circuit)

job = simulator.run(qobj)

#monitor job
job_monitor(job)

# get the result
result = job.result()

# get statevector
final_state = result.get_statevector()
plot_bloch_multivector(final_state)
```

Discussion
If the state of the target qubit after the SX gate was |0⟩, the Y-basis measurement will collapse it to either |Y+⟩ or |Y-⟩ with equal probability.
If the state of the target qubit after the SX gate was |1⟩ (the state we were interested in), the Y-basis measurement will collapse it to either |Y+⟩ or |Y-⟩, but with a phase difference due to the initial |1⟩ state. The phase difference will result in a relative phase between the |Y+⟩ and |Y-⟩ states, which means they won't have equal probabilities.
In other words, measuring the target qubit in the Y-basis introduces a phase difference that is based on the initial state of the qubit before the measurement. This phase difference will lead to differences in the measurement outcomes, making the probabilities of |Y+⟩ and |Y-⟩ unequal.

```python
# Building Circuit
circuit = QuantumCircuit(3,1)

# apply gate
circuit.h(0)
circuit.h(1)
circuit.ccx(0,1,2)
circuit.sx(2)

# apply measurement in y  basis
circuit.sdg(0)
circuit.h(0)







# drawing circuit
circuit.draw(output='mpl', initial_state=True)
```

```python

#Select a simulator
simulator = Aer.get_backend('aer_simulator')

#Save Unitary
circuit.save_unitary()

#assemble circuit
qobj = assemble(circuit)

#run on  simulator
job = simulator.run(qobj)

#monitor the process
job_monitor(job)

# get the result
result = job.result()

#get the unitary
unitary = result.get_unitary()

print(unitary)
array_to_latex(unitary)
```

Discussion: The difference in outcomes arises from the relative phase introduced by the Y-gate (SX gate) applied to the target qubit. Since the initial state of the target qubit was |1⟩, the Y-basis measurement will result in different probabilities for measuring |Y+⟩ and |Y-⟩. The relative phase between the basis states causes constructive or destructive interference, leading to different probabilities for the measurement outcomes.

In summary, the differences in outcome between Z-basis and Y-basis measurements arise from the different reference frames of measurement. Z-basis measurements are in the standard computational basis, while Y-basis measurements are in a rotated basis that introduces phase information. This phase information leads to variations in the probabilities of different measurement outcomes, even when the initial setup in the computational basis had equal probabilities.
